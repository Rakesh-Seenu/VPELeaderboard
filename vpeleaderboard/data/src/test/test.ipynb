```python
import yaml
import pandas as pd
# from vpeleaderboard.data.src.test import *

# train_duration: 10
# test_duration: 5

def run_model_for_time_period(model_id, duration):
    # Load the model
    load_biomodel(model_id)
    
    # Run the time course simulation
    results = run_time_course(duration)
    
    return results

def run_model_from_yaml(model_id, yaml_file):
    with open(yaml_file, 'r') as file:
        config = yaml.safe_load(file)
    
    train_duration = config['train_duration']
    test_duration = config['test_duration']
    
    # Run the model for training duration
    train_results = run_model_for_time_period(model_id, train_duration)
    
    # Run the model for testing duration
    test_results = run_model_for_time_period(model_id, test_duration)
    
    return train_results, test_results

# Example usage
model_id = 537  # Replace with your model ID
yaml_file = 'C:/Users/hadnesreenath/Documents/project/VPELeaderboard/vpeleaderboard/data/src/test/test.yaml'  # Replace with your YAML file path

train_results, test_results = run_model_from_yaml(model_id, yaml_file)

# Ensure the results are DataFrames
train_results_df = pd.DataFrame(train_results)
test_results_df = pd.DataFrame(test_results)
```
